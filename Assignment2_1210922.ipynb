**Unified version of the prices**

```python

# Custom function to calculate unified_price
def unified_price(row):
    #split unit column for each row 
    Value_unit= row['unit'].split()
    value = 1.0  
    if len(Value_unit) > 1 and Value_unit[0].isdigit():
        value = float(Value_unit[0]) 
    elif len(Value_unit) == 1 and Value_unit[0].isdigit():
        value = float(Value_unit[0])
        
    if 'KG' in row['unit']:
        return row['usdprice'] /value
    elif 'G' in row['unit']:
        return row['usdprice']  /(value/ 1000 )
    elif 'L' in row['unit']:
        return row['usdprice'] /value 
    elif 'Cubic meter' in row['unit']:
        return row['usdprice'] /(value * 1000 )
    else:
        return None
# Applying the function to each row
df_palestine_food_prices['unified_price'] = df_palestine_food_prices.apply(unified_price, axis=1)
df_palestine_food_prices
```

<span style="color:blue">This is the dataset after we calculated the unified price, so we created a new column called unified_price and the results were as follows: price in USD per kilogram, price per liter for each commodity measured by weight or volume

<span style="color:red"> **5. Do the prices effected by the number of violence events happened in the same month?interpret the results** </span>

```python
#split the date and convert type of month from int to string 
def month(row):
    Months_String= ["", "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    return Months_String[int(row['date'].split('/')[0])]
   
#create column month in df_palestine_food_prices    
df_palestine_food_prices['Month'] = df_palestine_food_prices.apply(month,axis=1)

df_palestine_food_prices['Month']
```

<span style="color:blue"> This series shows,new column month after split the date and convert the int month to string value </span>

```python
merge_df=pd.merge(df_palestine_food_prices,df_palestine_political_events,on="Month",how="left")
merge_df
```

<span style="color:blue"> This is the dataframe, after merging the two dataframe that have a common month</span>

```python
merge_df['unified_price'].corr(merge_df['Number of Events'])
```

<span style="color:blue"> According to the value of the correlation coefficient is **-0.00029801535601680653** close to zero and negative, it was found that prices are not affected by the number of violent events that happened in the same month.</span>

```python
merge_df.plot.scatter(x='Number of Events', y='unified_price');
```

<span style="color:blue"> According to the **scatter plot** representing the relationship between the number of violent events and prices, that the relationship is weak and prices are not affected by the number of violent events that happend in the same month.</span>
